```python
import numpy as np 
import pandas as pd
import seaborn as sns
import random
import json
import os
import functools

from matplotlib import pyplot as plt
from urllib.request import urlopen
from plotly import graph_objects as go
from plotly import express as px
```

```python
# Create mapbox token: https://docs.mapbox.com/help/getting-started/access-tokens/
mapbox_token = ''
px.set_mapbox_access_token(mapbox_token)
```

```python
path = 'drive/My Drive/place2crash_data/results/'
```

```python
def make_unique(IDs):
  mapped = {}
  unicount = 1
  IDs = IDs.to_list()

  for true_ID in IDs:
    if true_ID in mapped.keys():
      continue
    mapped[true_ID] = unicount
    unicount += 1

  for index, _ in enumerate(range(len(IDs))):
    IDs[index] = mapped[IDs[index]]

  return IDs

def load_plot(neighborhood = 'Harlem', min_size = 10, dist_threshold = 50, display_frame = True, zoom = 13.0):
  global cscale
  file_name = 'neighborhood' + '-' + neighborhood + '-' + str(dist_threshold) + '-cluster_info.csv'
  clusters = pd.read_csv(path + file_name, index_col = False)
  clusters = clusters[clusters['Cluster_Size'] > min_size]
  clusters = clusters.drop('ID_Neighborhood', axis = 1)
  clusters['Cluster ID'] = make_unique(clusters['Cluster ID'])
  
  if display_frame:
    display(clusters)

  print('', flush = True)

  fig = px.scatter_mapbox(clusters, lat = 'latitude', lon = 'longitude', opacity = 0.40,
                        color = 'Cluster ID', size = 'Degree', color_continuous_scale = 'hsv',
                        height = 680, size_max = 16, zoom = zoom, title = neighborhood)

  fig.show()
```

```python
load_plot(zoom = 12.75)
```
