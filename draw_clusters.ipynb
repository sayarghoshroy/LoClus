```python
import numpy as np 
import pandas as pd
import seaborn as sns
import random
import json
import os
import functools

from matplotlib import pyplot as plt
from urllib.request import urlopen
from plotly import graph_objects as go
from plotly import express as px
```

```python
# Create one here: https://docs.mapbox.com/help/getting-started/access-tokens/
mapbox_token = '_'
px.set_mapbox_access_token(mapbox_token)
```

```python
path = 'drive/My Drive/place2crash_data/results/'
```

```python
cscale = [[0.0, 'rgb(165,0,38)'], 
          [0.004, 'rgb(215,48,39)'], 
          [0.01, 'rgb(250, 152, 122)'], 
          [0.08, 'rgb(208, 254, 144)'], 
          [0.1, 'rgb(0, 255, 179)'], 
          [0.3, 'rgb(171,217,233)'], 
          [0.7, 'rgb(116,173,209)'], 
          [0.9, 'rgb(69,117,180)'], 
          [1.0, 'rgb(49,54,149)']]
```

```python
def load_plot(neighborhood = 'Harlem', min_size = 10, dist_threshold = 50, display_frame = True):
  global cscale
  file_name = 'neighborhood' + '-' + neighborhood + '-' + str(dist_threshold) + '-cluster_info.csv'
  clusters = pd.read_csv(path + file_name, index_col = False)

  if display_frame:
    display(clusters)

  print('', flush = True)

  fig = px.scatter_mapbox(clusters[clusters['Cluster_Size'] > 10], lat = 'latitude', lon = 'longitude',
                        color = 'Cluster ID', size = 'Degree', color_continuous_scale = cscale,
                        height = 680, size_max = 16, zoom = 12.5)

  fig.show()
```

```python
load_plot()
```

```python
load_plot('Bushwick')
```

```python
load_plot('Upper West Side')
```

```python
load_plot('Upper East Side', min_size = 20)
```

```python
load_plot('Bedford-Stuyvesant', min_size = 20)
```
